```
import pandas as pd
import numpy as np
from kmodes.kprototypes import KPrototypes
```

```
marketing_df = pd.read_csv('https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/marketing_cva_f.csv')
```

```
marketing_df.head()
```

```
marketing_df=marketing_df.drop(["Customer","Vehicle_Class","avg_vehicle_age","months_last_claim","Total_Claim_Amount"],axis=1)
```

```
marketing_df.head()
```

```
marketing_df.values
#convert pandas array to numpy array
```

```
mark_array= marketing_df.values
```

```
mark_array
```

```
mark_array[0:,1]=mark_array[0:,1].astype(float)
mark_array[0:,3]=mark_array[0:,3].astype(float)
mark_array[0:,5]=mark_array[0:,1].astype(float)
mark_array[0:,6]=mark_array[0:,1].astype(float)
```

```
mark_array
```

```
kproto = KPrototypes(n_clusters=3, verbose=2,max_iter=20)
clusters=kproto.fit_predict(mark_array,categorical=[0,2,4])
```

```
print(kproto.cluster_centroids_)
```

```
cluster_dict=[]

for c in clusters:
  cluster_dict.append(c)
```

```
cluster_dict
```

```
marketing_df['clusters']=cluster_dict
```

```
marketing_df
```
