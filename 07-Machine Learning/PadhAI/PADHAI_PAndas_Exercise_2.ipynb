```
def create_df(nRows,nCol,maxRand=10):
  arr = np.random.randint(0,maxRand,(nRows,nCol))
  df= pd.DataFrame(arr)
  df.index=['R' + str(x) for x in np.arange(1,nRows+1)]
  df.columns=['C'+ str(x) for x in np.arange(1, nCol+1)]
  return df
```
