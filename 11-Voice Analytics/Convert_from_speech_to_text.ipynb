```
import glob
def get_filenames(dirpath):
  fileset = [file for file in glob.glob(dirpath + "**/*.wav", recursive=True)]
  return fileset
```

```
path='/content/drive/My Drive/Colab Notebooks/Dataset/Speak_Recog_Data/Audio/'
list_of_files =get_filenames(path)
```

```
def audio_to_text(filepath):
  # open the file
  r = sr.Recognizer()
  text=""
  with sr.AudioFile(filepath) as source:
    # listen for the data (load audio to memory)
    audio_data = r.record(source)
    # recognize (convert from speech to text)
    text = r.recognize_google(audio_data)
  return text
```

```
# Create Text dictonaries
number_of_files=len(filepathlist)
audio_texts_dict = {}
if( number_of_files>1 ):
  for filename in filepathlist:
    print(filename)
    text=audio_to_text(filename)
    file_name=(os.path.splitext(os.path.basename(filename))[0])
    print(file_name)
    print(text)
    if filename not in audio_texts_dict:
      audio_texts_dict[file_name] = []
      audio_texts_dict[file_name].append(text)
```
