```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
!wget https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/Images/367bb0088a3c78036c77ff3fac3340c0.jpg
```

```python
image = cv2.imread('367bb0088a3c78036c77ff3fac3340c0.jpg', flags=cv2.IMREAD_COLOR)
```

```python
cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_blurred = cv2.blur(src=image, ksize=(5, 5))
cv2_imshow(image_blurred)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_blurred = cv2.blur(src=image, ksize=(75, 75))
cv2_imshow(image_blurred)
cv2.waitKey()
cv2.destroyAllWindows()
```
