```python
# import the required packages
import cv2
import os
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
!wget https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/Images/Nature.jpg
```

```python
# load the image into system memory
image = cv2.imread('Nature.jpg', flags=cv2.IMREAD_COLOR)
cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print(image.shape)
print("Rows of Pixels: %d Rows"%(image.shape[0]))
print("Columns of Pixels: %d Columns"%(image.shape[1]))
print("Color Channels: %d Color Channels"%(image.shape[2]))
```

```python
import numpy as np
numpy_zeros = np.zeros(shape=(6, 6))
print(numpy_zeros)
```

```python
numpy_ones = np.ones(shape=(4, 4))
print(numpy_ones)
```

```python
numpy_zeros[1:5, 1:5] = numpy_ones
print(numpy_zeros)
```

```python
tree = image[320:393, 570:665]
cv2_imshow(tree)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
print('Shape Of Tree: ', tree.shape)
print("Rows of Pixels: %d Rows"%(tree.shape[0]))
print("Columns of Pixels: %d Columns"%(tree.shape[1]))
print("Color Channels: %d Color Channels"%(tree.shape[2]))
```

```python
image[333:406, 0:87] = tree
image[333:406, 87:174] = tree[:, ::-1, :]
image[323:396, 174:261] = tree[:, ::-1, :]
image[323:396, 267:342] = tree[:, :, :]
image[312:385, 342:429] = tree[:, ::-1, :]
image[315:388, 429:516] = tree
```

```python
cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
cv2_imshow('Cropped Trees', image)
cv2.waitKey()
cv2.destroyAllWindows()
```
