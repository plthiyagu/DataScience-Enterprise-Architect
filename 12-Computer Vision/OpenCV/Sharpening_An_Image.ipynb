```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
!wget https://raw.githubusercontent.com/plthiyagu/Personnel/master/Dataset/Images/367bb0088a3c78036c77ff3fac3340c0.jpg
```

```python
image = cv2.imread('367bb0088a3c78036c77ff3fac3340c0.jpg', flags=cv2.IMREAD_COLOR)
```

```python
# display the image to the screen
cv2_imshow(image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])
image_sharp = cv2.filter2D(src=image, ddepth=-1, kernel=kernel)
cv2_imshow(image_sharp)
cv2.waitKey()
cv2.destroyAllWindows()
```
