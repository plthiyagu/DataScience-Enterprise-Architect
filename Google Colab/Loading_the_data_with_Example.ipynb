```
movies_sheet1 = pd.read_excel(movie_file_name, sheet_name=0, index_col=0)
movies_sheet1.head()
```

```
movies_sheet2 = pd.read_excel(movie_file_name, sheet_name=1, index_col=0)
movies_sheet2.head()
```

```
movies_sheet3 = pd.read_excel(movie_file_name, sheet_name=2, index_col=0)
movies_sheet3.head()
```

```
movies = pd.concat([movies_sheet1, movies_sheet2, movies_sheet3])
movies.head()
movies.shape
```

```
xlsx = pd.ExcelFile(movie_file_name)
movies_sheets = []
for sheet in xlsx.sheet_names:
    movies_sheets.append(xlsx.parse(sheet))
    movies = pd.concat(movies_sheets)
```

```
movies.head()
```

```
movies.to_excel('output.xlsx', index=False)
```
