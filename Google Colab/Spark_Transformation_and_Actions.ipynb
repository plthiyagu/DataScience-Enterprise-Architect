```
import pyspark.sql.functions as F
corana_df.groupBy("Country","State_cleaned").agg(F.max("Date")).show()
```

```
corana_df.join(corana_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=['Country','State_cleaned','Date'],how="inner").show()
```

```
corana_df.join(corana_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=['Country','State_cleaned','Date'],how="inner").sort("Confirmed",ascending=False).show()
```

```
from pyspark.sql.functions import row_number
from pyspark.sql.window import Window

ws=Window().partitionBy("Country","State_cleaned").orderBy(col('Date').desc())
corana_df.withColumn("row_num",row_number().over(ws)).show()
```

```
corana_df.withColumn("row_num",row_number().over(ws)).where(col("row_num")==1).show()
```

```
corona_max_df=corana_df.join(corana_df.groupBy("Country","State_cleaned").agg(F.max("Date").alias("Date")),on=['Country','State_cleaned','Date'],how="inner").sort("Confirmed",ascending=False)
```
